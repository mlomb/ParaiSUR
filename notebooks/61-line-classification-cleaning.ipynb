```python
with open("professional_lines.txt", "r") as f:
    professional_lines = f.read().splitlines()

with open("not_professional_lines.txt", "r") as f:
    not_professional_lines = f.read().splitlines()

len(professional_lines), len(not_professional_lines)
```

```python
professional_lines_clean = []
not_professional_lines_clean = []

for line in not_professional_lines:
    EXCLUDE = [
        "Fees",
        "Travel Expense",
        "Travei Expense",
        "Amount Overdue",
        "Client meeting",
        "Client workshop",
        "Indivdual",
        "Group Meal",
        "2020-", "2021-", "2022-", "2023-",
        "Equipment",
        "Credit note"
    ]

    if not any([e.lower() in line.lower() for e in EXCLUDE]):
        not_professional_lines_clean.append(line)

for line in professional_lines:
    EXCLUDE = [
        "Time entered",
        "Engineer",
        "Design",
        "Manager",
        "Consultant",
        "Consutlant",
        "Specialist",
        "ELECTRICAL",
        "Sr.",
        "Supervisor"
    ]

    if not any([e.lower() in line.lower() for e in EXCLUDE]):
        professional_lines_clean.append(line)
```

```python
len(professional_lines_clean), len(not_professional_lines_clean)
```

```python
# save list to files
with open("professional_lines_clean.txt", "w") as f:
    for line in professional_lines_clean:
        f.write(line + "\n")

with open("not_professional_lines_clean.txt", "w") as f:
    for line in not_professional_lines_clean:
        f.write(line + "\n")
```

```python
from lib.data import load_first_names, load_last_names


names = load_first_names()
lasts = load_last_names()

for line in not_professional_lines:
    for name in names:
        if name.lower() in line.lower():
            for last in lasts:
                words = line.split()
                for word in words:
                    if (name + last).lower() == word:
                        print(name, last, line)
```

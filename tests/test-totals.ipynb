```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df_submit = pd.read_csv("../submissions/48_use_truth.csv")
df_val = pd.read_csv("./validation.csv")
df_val["True"] = df_val["Total Charged"]
```

```python
df_submit, df_val
```

```python
# join dataframes
df = pd.merge(df_submit, df_val, on='Invoice Number')
df["diff"] = df["True"] - df["Total Charged_x"]
df.sort_values(by="diff")
```

```python
# MAE
df["diff"].abs().mean()
```
